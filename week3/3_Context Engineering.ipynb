# Prompt engineering 可以讓回答更漂亮，但無法提供模型沒有的資訊。

```python
from google.colab import userdata
from openai import OpenAI
import json
import pandas as pd
from IPython.display import display, Markdown

# 從 Colab Secrets 讀取
GROQ_API_KEY = userdata.get("GROQ_API_KEY")

# Groq 為 OpenAI 相容 API
client = OpenAI(
    api_key=GROQ_API_KEY,
    base_url="https://api.groq.com/openai/v1"
)

MODEL_NAME = "llama-3.3-70b-versatile"

def ask_llm(messages, model=MODEL_NAME, temperature=0.2):
    resp = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature
    )
    return resp.choices[0].message.content
```

```python
messages = [
    {"role": "user", "content": "請根據 W 公司 2025 年永續報告，說明最大的碳排來源與改善措施。"}
]
print(ask_llm(messages))
```

```python
messages = [
    {"role": "system", "content": "你是一位專業永續顧問，請準確、條列、專業地回答。"},
    {"role": "user", "content": "請根據 W 公司 2025 年永續報告，說明最大的碳排來源與改善措施，並分成三點說明。"}
]
print(ask_llm(messages))
```

## 加入文件 Context，模型才真的有依據

```python
document = """
W 公司 2025 年永續報告摘要：
1. 本公司碳排放最大來源為範疇二，主要來自外購電力。
2. 今年已導入能源管理系統與高效空壓機，預估節電 8%。
3. 未來將推動再生能源導入與老舊設備汰換。
"""

messages = [
    {
        "role": "system",
        "content": "你是一位永續分析助理，只能根據提供的文件回答；若文件沒提到，請明確說明。"
    },
    {
        "role": "user",
        "content": f"文件如下：\n{document}\n\n請根據文件，說明 W 公司最大的碳排來源與改善措施。"
    }
]
print(ask_llm(messages))
```

### Context engineering 不是「多放資料」就好，而是要放對資料。

```python
wrong_document = """
W 公司 2025 年永續報告摘要：
1. 本公司碳排放最大來源為範疇一，主要來自公司公務車燃油。
2. 改善措施以植樹活動為主。
"""

messages = [
    {
        "role": "system",
        "content": "你是一位永續分析助理，只能根據提供的文件回答。"
    },
    {
        "role": "user",
        "content": f"文件如下：\n{wrong_document}\n\n請根據文件，說明 W 公司最大的碳排來源與改善措施。"
    }
]
print(ask_llm(messages))
```

### context engineering 的價值，不只是塞文件，而是決定模型在回答前，回答問題所需要的重點資訊

```python
conversation = [
    "使用者：我正在設計一門給設計學院學生的生成式 AI 課程。",
    "助理：了解，這門課偏理論還是偏實作？",
    "使用者：我希望偏實作，但也想讓學生知道大型語言模型的核心概念。",
    "助理：學生有程式基礎嗎？",
    "使用者：大部分沒有，所以我不想放太多數學。",
    "助理：那適合用圖像化、互動化的方式教學。",
    "使用者：對，而且我希望能用 Colab 做一些簡單示範。",
    "助理：明白，這樣可以降低環境安裝門檻。",
    "使用者：另外我也想讓學生知道 prompt engineering 和 context engineering 的差別。",
    "助理：這很重要，可以用對照實驗來呈現。",
    "使用者：最後我希望教材可以連到 RAG 和 Agent 的概念。",
]

def compress_context(conversation_list, model=MODEL_NAME):
    conversation_text = "\n".join(conversation_list)

    messages = [
        {
            "role": "system",
            "content": (
                "你是一個專門做 context compression 的 AI 助手。\n"
                "請將對話壓縮成『後續回答仍需要的重要記憶』。\n"
                "請遵守以下規則：\n"
                "1. 只保留對未來任務有幫助的資訊\n"
                "2. 移除寒暄、重複、無關內容\n"
                "3. 保留使用者目標、偏好、限制條件、任務背景\n"
                "4. 使用繁體中文\n"
                "5. 以條列式輸出，最多 8 點\n"
            )
        },
        {
            "role": "user",
            "content": f"請壓縮以下對話：\n\n{conversation_text}"
        }
    ]

    resp = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0.1
    )

    return resp.choices[0].message.content


compressed_memory = compress_context(conversation)
print(compressed_memory)
```

```python
new_question = "請幫我設計第一堂課的開場活動"

raw_messages = [
    {"role": "system", "content": "你是一位教學設計顧問，請用繁體中文回答。"},
    {"role": "user", "content": "\n".join(conversation)},
    {"role": "user", "content": new_question}
]

raw_resp = client.chat.completions.create(
    model=MODEL_NAME,
    messages=raw_messages,
    temperature=0.3
)

print("=== 使用原始長對話 ===")
print(raw_resp.choices[0].message.content)
```

```python
summary_messages = [
    {"role": "system", "content": "你是一位教學設計顧問，請用繁體中文回答。"},
    {"role": "user", "content": f"以下是使用者的重要背景摘要：\n{compressed_memory}"},
    {"role": "user", "content": new_question}
]

summary_resp = client.chat.completions.create(
    model=MODEL_NAME,
    messages=summary_messages,
    temperature=0.3
)

print("=== 使用壓縮後摘要 ===")
print(summary_resp.choices[0].message.content)
```
